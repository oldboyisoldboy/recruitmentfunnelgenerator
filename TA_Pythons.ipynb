```python
#@title Recruitment Funnel V2

print("Welcome to your Recruitment Funnel Generator")
print()

class Stat: #OBJECT
  def __init__(self,total_mrf, sourced_applicants, failed_cvscreening, invited_hrscreening, hr_screened, endorsed_dept, show_up, passed, hired): #CONSTRUCTOR
    self.total_mrf = total_mrf
    self.sourced_applicants = sourced_applicants
    self.failed_cvscreening = failed_cvscreening
    self.invited_hrscreening = invited_hrscreening
    self.hr_screened = hr_screened
    self.endorsed_dept = endorsed_dept
    self.show_up = show_up
    self.passed = passed
    self.hired = hired

  def funnel(self): #OBJECT FUNCTION
    print ("This is the total number HC in your MRF :" + str(self.total_mrf))
    print ("The Number of your Sourced Applicants is  :" + str(self.sourced_applicants))
    print ("This is the number those who the failed CV Screening :" + str(self.failed_cvscreening))
    print ("Here is the number of those who are Invited for HR Screening :" + str(self.invited_hrscreening))
    print ("This is the number of applicants interviewed by HR :" + str(self.hr_screened))
    print ("Total number of endorsed applicants for Department Interview :" + str(self.endorsed_dept))
    print ("Applicants who Showed Up :" + str(self.show_up))
    print ("Total number of those who passed the Final Interview :" + str(self.passed))
    print ("Number of Hires Made to date :" + str(self.hired))

recruitment_numbers = []

while True:
    total_mrf = int(input ("Please enter total number of headcount in MRF :"))
    sourced_applicants = int(input("Please enter total number of Sourced Applicants  :"))
    failed_cvscreening = int(input("Please enter total number of failed CV Screening :"))
    invited_hrscreening = int(input("Please enter total number of Invited for HR Screening :"))
    hr_screened = int(input("Please enter total number of HR Screened :"))
    endorsed_dept = int(input("Please enter total number of Endorsed to Department :"))
    show_up = int(input("Please enter total number of Show Up :"))
    passed = int(input("Please enter total number of Passed Final Interview :"))
    hired = int(input("Please enter numnber of Hires Made :"))
    s = Stat (total_mrf,sourced_applicants,failed_cvscreening,invited_hrscreening,hr_screened,endorsed_dept,show_up,passed,hired)
    recruitment_numbers.append(s)

    print("\n")
    choice = input ("Would you like to edit your funnel? [Y/N] : ")
    if choice == "Y" or choice == "y" : pass
    else:
      break
    print("\n")
    print ("Calculating...")

for number in recruitment_numbers:
    print()
    print("This is your Recruitment Funnel: ")
    print()
    number.funnel()
  
def main():
  funnel1 = (round(invited_hrscreening / sourced_applicants *100))
  funnel2 = (round(failed_cvscreening / sourced_applicants * 100))
  funnelA = (round(hr_screened / invited_hrscreening * 100))
  funnelB = (round(endorsed_dept / hr_screened * 100))
  funnelC = (round(show_up / endorsed_dept *100))
  funnelD = (round(passed / show_up * 100))
  funnelE = (round(hired / passed * 100))
  funnelF = (round(hired / total_mrf *100))

  print ("Your current headcount requirement for this MRF is " + str(total_mrf) + ".")
  print ("Out of the " + str(sourced_applicants) + " applications received, " + str(invited_hrscreening) + ' candidates were invited for HR initial screening. Inviting out ' + str(funnel1) + "%" + ' of the applicants.')
  print (str(failed_cvscreening) + " applicants failed the CV screening and is at " + str(funnel2)+"% loss in conversion. This could be due to various factors.")
  print ('However, only ' + str(hr_screened) + " of them attended the HR initial screening. Giving us a " + str(funnelA)+ "%" + " show-up rate for TA interviews.")
  print ('From the ' + str(hr_screened) + " applicants screened, " + str(endorsed_dept) + ' were endorsed for department screening. ' + str(funnelB)+"%" + " conversion rate from HR interview to hiring department interview.")
  print (str(show_up) + ", or " + str(funnelC)+'%' + ' of which showed up and attended the final interview.')
  print ("The hiring department selected " + str(passed) + " applicants after the assessments. A passing rate of " + str(funnelD)+"%" + " from the endorsed candidates.")
  print ('At present, we were able to onboard ' + str(hired) + ' applicants from those who passed the interviews, furthering our conversion to ' + str(funnelE)+'%' + " from passing the final interview to closing vacancy.")
  print ('')
  print ("Looking at your at your total HC which is " +str(total_mrf) + " you are at " + str(funnelF) + "% from completing the Manpower Request.")

print('')
print('Here is your funnel report:')
print('')

main()


```
