```python
#@title Recruitment Funnel Generator. Kindly click the play button beside show code.
print('Welcome to TA Funnel Generator!')
print("For now, 'zero' is not accepted by this generator.")
print('')

#Stats

while True:
  total_mrf = int(input("Please enter the total number of HC for your MRF: "))
  if total_mrf <= 0:
     print ('Please input a number greater than 0')
  elif total_mrf >=1:
    break

while True: 
  sourced_applicants = int(input("Please enter the number of your Sourced Applicants: "))
  if sourced_applicants <=0:
    print ('Please input a number greater than 0')
  elif sourced_applicants >=1:
    break

while True:
  failed_cvscreening = int(input("Please enter the number of those who failed CV screening: "))
  if failed_cvscreening <= 0:
    print ('Please input a number greater than 0')
  elif sourced_applicants >=1:
    break

while True:
  invited_hrscreening = int(input("Please enter the number of invited for HR Screening: "))
  if invited_hrscreening <=0:
    print ('Please input a number greater than 0')
  else:
    break

while True:
  hr_screened = int(input("Please enter the number of HR Screened: "))
  if hr_screened <= 0:
    print ('Please input a number greater than 0')
  else:
    break

while True:
  endorsed_dept = int(input("Please enter the number endorsed to department: "))
  if endorsed_dept <= 0:
    print ('Please input a number greater than 0')
  else:
    break

while True:
  show_up = int(input("Please enter the number of Show Up: "))
  if show_up <= 0:
    print ('Please input a number greater than 0')
  else:
    break

while True:
  passed = int(input("Please enter the number of Passed: "))
  if passed <= 0:
    print ('Please input a number greater than 0')
  else:
    break

while True:
  hired = int(input("Please enter the number of Hires Made: "))
  if hired <= 0:
    print ('Please input a number greater than 0')
  else:
    break

def main():
  funnel1 = (round(invited_hrscreening / sourced_applicants *100))
  funnel2 = (round(failed_cvscreening / sourced_applicants * 100))
  funnelA = (round(hr_screened / invited_hrscreening * 100))
  funnelB = (round(endorsed_dept / hr_screened * 100))
  funnelC = (round(show_up / endorsed_dept *100))
  funnelD = (round(passed / show_up * 100))
  funnelE = (round(hired / passed * 100))
  funnelF = (round(hired / total_mrf *100))

  print ("Your current headcount requirement for this MRF is " + str(total_mrf) + ".")
  print ("Out of the " + str(sourced_applicants) + " applications received, " + str(invited_hrscreening) + ' candidates were invited for HR initial screening. Inviting out ' + str(funnel1) + "%" + ' of the applicants.')
  print (str(failed_cvscreening) + " applicants failed the CV screening and is at " + str(funnel2)+"% loss in conversion. This could be due to various factors.")
  print ('However, only ' + str(hr_screened) + " of them attended the HR initial screening. Giving us a " + str(funnelA)+ "%" + " show-up rate for TA interviews.")
  print ('From the ' + str(hr_screened) + " applicants screened, " + str(endorsed_dept) + ' were endorsed for department screening. ' + str(funnelB)+"%" + " conversion rate from HR interview to hiring department interview.")
  print (str(show_up) + ", or " + str(funnelC)+'%' + ' of which showed up and attended the final interview.')
  print ("The hiring department selected " + str(passed) + " applicants after the assessments. A passing rate of " + str(funnelD)+"%" + " from the endorsed candidates.")
  print ('At present, we were able to onboard ' + str(hired) + ' applicants from those who passed the interviews, furthering our conversion to ' + str(funnelE)+'%' + " from passing the final interview to closing vacancy.")
  print ('')
  print ("Looking at your total HC which is " +str(total_mrf) + " you are at " + str(funnelF) + "% from completing the Manpower Request.")


print('')
print('Here is your funnel report:')
print('')

main()
```
